```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
urlbase = 'node_modules/cesium/Build/Cesium/Assets/Textures/NaturalEarthII'
```

```python
cols = []

for icol in range(8):
    tiles = []
    for i in range(4):
        file = urlbase+'/2/{}/{}.jpg'.format(icol,3-i)
        tiles.append(plt.imread(file))
    col = np.vstack(tiles)
    cols.append(col)
```

```python
image = np.hstack(cols)
```

```python
plt.imshow(image)
```

```python
plt.imsave('imagen.png', image)
```
